# Scrapping Reviews & Ratings of Phones from Flipkart

```python
#Activating the chrome browser
driver=webdriver.Chrome(r"C:\Users\MUNGI\Downloads\chromedriver_win32 (1)\chromedriver.exe")

#Opening homepage of flipkart.com
driver.get('https://www.flipkart.com')
#Asking the user to input the keywords to search
user_inp=input('Enter the product you want to search : ')
```

```python
#Going to next page
cross=driver.find_element_by_xpath('//button[@class="_2KpZ6l _2doB4z"]')
cross.click()
```

```python
#Entering product in search bar and clicking on the search button
search_bar=driver.find_element_by_xpath("//input[@class='_3704LK']")  
search_bar.clear()                                            
search_bar.send_keys(user_inp)                               
time.sleep(4)
search_button=driver.find_element_by_xpath("//button[@class='L0Z3Pu']") 
search_button.click()
```

```python
#Fetching urls from the first page
urls=[]
for i in driver.find_elements_by_xpath("//a[@class='_1fQZEK']"):
    urls.append(i.get_attribute("href"))
```

```python
#Displaying urls
urls
```

```python
#Length of urls
len(urls)
```

```python
#Defining empty list Variables
re=[]
ra=[]
```

```python
#Fetching data from the urls of first page
for i in urls:
    driver.get(i)
    time.sleep(5)
    
    driver.find_element_by_xpath("//div[@class='col JOpGWq']/a").click()
    time.sleep(2)
    
    #Fetching review
    try:
        r=driver.find_elements_by_xpath("//div[@class='t-ZTKy']/div/div")
        for j in r:
            re.append(j.text.replace("\n"," "))
    except NoSuchElementException as e:
        for j in r:
            re.append('-')
        
              

    #Fetching rating
    try:
        r1=driver.find_elements_by_xpath("//div[@class='col _2wzgFH K0kLPL']/div[1]/div[1]")
        for k in r1:
            ra.append(k.text.replace("\n"," "))
    except NoSuchElementException as e:
            for k in r1:
                ra.append('-')
                
```

```python
#Length of review and rating scrapped
print(len(re),len(ra))
```

```python
#Displaying reviews
re
```

```python
#Displaying ratings
ra
```

```python
#Going to next page
next=driver.find_element_by_xpath('//a[@class="_1LKTO3"]')
next.click()
```

```python
#Creating empty list to store urls of next pages
urls1=[]
#Fetching urls of first three pages
for i in range(1,4):
    for j in driver.find_elements_by_xpath("//a[@class='_1fQZEK']"):
        urls1.append(j.get_attribute("href"))
        driver.find_element_by_xpath('//a[@class="_1LKTO3"][2]')
        time.sleep(3)
```

```python
#Displaying urls of next pages retrieved
urls1
```

```python
#Length of urls retrieved
len(urls1)
```

```python
#Defining empty list Variables for data of next pages
re1=[]
ra1=[]
```

```python
#Fetching data from urls of next pages
for i in urls1:
    driver.get(i)
    time.sleep(4)
    
    driver.find_element_by_xpath("//div[@class='col JOpGWq']/a").click()
    time.sleep(2)
    
    #Fetching review
    try:
        r2=driver.find_elements_by_xpath("//div[@class='t-ZTKy']/div/div")
        for j in r2:
            re1.append(j.text.replace("\n"," "))
    except NoSuchElementException as e:
        for j in r2:
            re1.append('-')
        
              

    #Fetching rating
    try:
        r3=driver.find_elements_by_xpath("//div[@class='col _2wzgFH K0kLPL']/div[1]/div[1]")
        for k in r3:
            ra1.append(k.text.replace("\n"," "))
    except NoSuchElementException as e:
            for k in r3:
                ra1.append('-')
                
```

```python
#Length of data scrapped
print(len(re1),len(ra1))
```

```python
#Closing web driver
driver.close()
```

```python
#Creating dataframe 1
rp1=pd.DataFrame({})
rp1['Reviews']=re
rp1['Ratings']=ra
```

```python
#Creating dataframe 2
rp2=pd.DataFrame({})
rp2['Reviews']=re1
rp2['Ratings']=ra1
```

```python
#Concatenation of 2 dataframes = Total data scrapped from flipkart for phones
rp=pd.concat([rp1,rp2],ignore_index=True)
rp
```

```python
#Entering product in search bar and clicking on the search button
search_bar=driver.find_element_by_id("twotabsearchtextbox")   #Locating search_bar by id
search_bar.clear()                                            #Clearing search bar
search_bar.send_keys(user_inp)                                #sending user input to search bar
time.sleep(4)
search_button=driver.find_element_by_xpath('//div[@class="nav-search-submit nav-sprite"]')   #Locating search bar by id
search_button.click()
```

```python
#Fetching urls
url5=[]
for i in driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']/a"):
    url5.append(i.get_attribute("href"))
```

```python
#Length of urls 
len(url5)
```

```python
#Creating empty list 
re_7 = []
ra_7 = []


for i in url5[0:2]:
    driver.get(i)
    time.sleep(3)
 
    
    #clicking on see all review
    driver.find_element_by_xpath("//a[@class='a-link-emphasis a-text-bold']").click()
    time.sleep(3)

    #Scrape reviews
    while True:
        
        try:
            #scrape reviews
            c=driver.find_elements_by_xpath('//div[@class="a-row a-spacing-small review-data"]//span//span')  # scraping reviews
            for m in c:
                re_7.append(m.text.replace("\n"," "))         
        except:
            break
        
        #scrape ratings
        try:
            # scrape ratings
            b=driver.find_elements_by_xpath('//i[@data-hook="review-star-rating"]')  # scraping ratings
            for l in b:
                ra_7.append(l.get_attribute("class"))
            
            driver.find_element_by_xpath('//ul[@class="a-pagination"]/li[2]/a').click()  # clicking on next button
            time.sleep(3)
        except:
            break
```

```python
#Closing web driver
driver.close()
```

```python
#Length of data scraped
print(len(re_7),len(ra_7))
```

```python
#Creating dataframe for Reviews
rat2=pd.DataFrame({})
rat2['Reviews']=re_7
rat2
```

```python
#Creating dataframe for Ratings
rat3=pd.DataFrame({})
rat3['Ratings']=ra_7
```

```python
#dropping unnecessary extra rows from the column 
rat2.drop(np.arange(2296,2379), axis=0, inplace=True)
```

```python
#Merging of two dataframes into to achieve scrapped info
amazon2 = pd.concat([rat2, rat3], axis=1)
```

```python
#Extracting number(ratings) from scrapped data
def number(n):
    num = re.findall(r'[0-9]+',n)
    return "".join(num)
amazon2['Ratings']=amazon2['Ratings'].apply(lambda x: number(x))
```

# Reviews and Ratings from Flipkart for Phones & Earphones

### Phones

```python
#Activating the chrome browser
driver=webdriver.Chrome(r"C:\Users\MUNGI\Downloads\chromedriver_win32 (1)\chromedriver.exe")

#Opening homepage of flipkart.com
driver.get('https://www.flipkart.com/i-kall-k11/product-reviews/itmez7zmwmrsgrfg?pid=MOBG56SUEEA6QGZ6&lid=LSTMOBG56SUEEA6QGZ63HJCTA&marketplace=FLIPKART&page=2')
```

```python
#Creating empty list to fill info
re_8=[]
ra_8=[]
```

```python
#Fetching data from the urls of first page
for i in range(1,118):
    time.sleep(2)
    
    try:
        #scrape reviews
        x=driver.find_elements_by_xpath('//div[@class="col _2wzgFH K0kLPL"]/div[2]/div[1]/div[1]/div[1]')  # scraping reviews
        for m in x:
            re_8.append(m.text.replace("\n"," "))
    except NoSuchElementException as e:
        for m in x:
            re_8.append('-')    
        
            
    #scrape ratings
    try:
        # scrape ratings
        y=driver.find_elements_by_xpath('//div[@class="col _2wzgFH K0kLPL"]/div[1]/div[1]')  # scraping ratings
        for n in y:
            ra_8.append(n.text.replace("\n"," "))
    except NoSuchElementException as e:
        for n in y:
            ra_8.append('-')
            
            
    driver.find_element_by_xpath('//a[@class="_1LKTO3"][2]').click()  # clicking on next button
    time.sleep(3)
    
```
